##### **Observation**

1. The columns names needs to be standardised
2. There are values filled as '..', probably missing values. Each record is a country KPI showing stats trend over the year. These '..' can be replaced with 0
3. All columns are objects datatype. The stats columns needs to be converted to numeric

```python
#remove the spaces. convert caps to lower case and strip the duplicated naming on year columns i.e 2015 [YR2016]
#to yr2016
columns1 = list(mobile_df.columns[0:4].str.lower().str.replace(' ', '_'))
columns2 = list(mobile_df.columns[4:].str.slice(start=6, stop=12, ).str.lower())
mobile_df.columns = columns1 + columns2
```

```python
mobile_df.head()
```

```python
#replace the '..' values in stats columns yr1995 to yr2016 with 0 and convert the columns to int type
#pick columns of interest
yearly_columns = mobile_df.columns[4:]
mobile_df[yearly_columns] = mobile_df[yearly_columns].apply(lambda x: x.replace('..', '0')).astype(float).astype('int')

#drop column yr2016 with only zeros
mobile_df.drop(columns=['series_code', 'country_code','yr2016'], inplace=True)
mobile_df.sample(10)
```

```python
stats_columns = list(mobile_df.columns[1:])
trade_trend_df = mobile_df.groupby('series_name', as_index=False).sum()
trade_trend_df = trade_trend_df.transpose().reset_index()
trade_trend_df
new_columns = trade_trend_df.iloc[0]
new_columns
trade_trend_df.columns = new_columns
trade_trend_df.drop(index=0, inplace=True)
trade_trend_df
value_columns = trade_trend_df.columns[1:]
trade_trend_df = pd.melt(trade_trend_df, id_vars=['series_name'], value_vars=value_columns)


#ploting the graph
# Render
fig = px.line(
    trade_trend_df, 
    x = 'series_name',
    y = 'value',
    color = 0,

    template = "simple_white"
  )

# Tweak and Label
fig.update_layout(
    title = {
        'text': 'Major KPIs trend over the years', 
        'x': 0.5,
        'font_size' : 15,
        'xanchor': 'center',
        'yanchor': 'top'},
    xaxis_title_text = 'Years',
    yaxis_title_text = 'amount',
  ) 

# Display
fig.show()
```

```python
yearly_columns = list(mobile_df.columns[2:])
country_trend_df = pd.melt(mobile_df, id_vars=['series_name', 'country_name'], value_vars=yearly_columns)
cellular_df = country_trend_df[country_trend_df['series_name']=='Fixed broadband subscriptions']

fig = px.line(
    cellular_df, 
    x = 'variable',
    y = 'value',
    color = 'country_name',

    template = "simple_white"
  )

# Tweak and Label
fig.update_layout(
    title = {
        'text': 'fixed broadband subscriptions growth in numbers per country', 
        'x': 0.5,
        'font_size' : 15,
        'xanchor': 'center',
        'yanchor': 'top'},
    xaxis_title_text = 'Years',
    yaxis_title_text = 'fixed broadband subscriptions',
  ) 

# Display
fig.show()
```
